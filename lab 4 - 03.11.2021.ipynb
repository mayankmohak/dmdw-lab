Q3) Lets assume pre = np.array([19, 6, 11, 9]) and post = np.array([15, 11, 9, 8]) are the two
Numpy array, plot this two array as side by side bar plot. 

```python
import numpy as np
from matplotlib import pyplot as plt

pre = np.array([19,6,11,9])
post = np.array([15,11,9,8])
labels = ['Survey'+ x for x in list('ABCD')]
fig,ax = plt.subplots(figsize=(8,3.5))
width = 0.4
xlocs  = np.arange(len(pre))
ax.bar(xlocs,pre,width,color='green',label='True')
ax.bar(xlocs,post,width,color='blue',label='False')
ax.set_xticks(ticks=range(len(pre)))
ax.set_xticklabels(labels)
ax.legend(loc='best')
ax.set_ylabel('count')
fig.suptitle('sample chart')
fig.savefig('chart.png',dpi=125)
```

Q4) Generate three NumPy arrays (X1, X2, X3) by getting 20 equidistance points between 0-
100. Next convert this array into Y1, Y2, Y3 by applying sin, cos, and tan function over X1,
X2, X3, respectively. Finally, plot these multiple lines over the same axis on a line plot graph.

```python
import numpy as np
from matplotlib import pyplot as plt

x1=np.linspace(0,100,20)
x2=np.linspace(0,100,20)
x3=np.linspace(0,100,20)
y1 = np.sin(x1)
y2 = np.cos(x2)
y3 = np.tan(x3)

fig,ax = plt.subplots()

ax.plot(x1,y1,label='sin')
ax.plot(x2,y2,label='cos')
ax.plot(x3,y3,label='tan')
ax.legend(loc='best',prop={'size':'large'})
fig.suptitle('sample chart')
fig.savefig('chart1.png',dpi=125)
```

Q5) Write a python program: Create list x = [5, 2, 9, 4, 7] and y = [10, 5, 8, 4, 2] : perform line,
scatter functions over the list x and y. b) Plotting two or more lines on the same plot
Y1=[8,6,7,5,1]. c) Naming the x and y-axis. Giving a title to my graph. Show a legend on the
plot.

```python
import numpy as np
from matplotlib import pyplot as plt

x =[5,2,9,4,7]
y =[10,5,8,4,2]
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
y1=[8,6,7,5,1]
plt.plot(x,y,label='line 1')
plt.plot(x,y1,label='line 2')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Two or more lines or same Plot')
plt.legend()
plt.show()
```
